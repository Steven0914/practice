```python
# 필요없는 데이터 삭제(이름 이메일)
X = car_df.drop(['Customer Name','Customer e-mail','Country','Car Purchase Amount'],axis=1)
X
```

```python
y = car_df['Car Purchase Amount']
y
```

```python
# 정규화해줘야함
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
# 최대값과 최소값을 기준으로 0과 1사이로 표준화함 
# 실제 데이터의 최대값과 최소값을 기준으로 표준화가 아님 데이터보다 작은값과 큰값이 있음

X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
scaler.data_max_
```

```python
scaler.data_min_
```

```python
y= y.values.reshape(-1,1)
# scaler를 사용하려면 2차원 데이터 구조여야함
# 그래서 reshape(-1.1)을 하면 -1은 자동으로 차원 크기계산하고 1차원 하니까 (500,1)이됨

y_scaled = scaler.fit_transform(y)
# 이코드는 데이터의 평균과 표준편차를 구하고, 이를 이용해 데이터를 정규화해줌
```

# 모델 학습

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled,y_scaled)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled,y_scaled,test_size=0.15)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
import tensorflow.keras
from keras.models import Sequential # 신경망을 순차적으로 설계해나가는 모델
from keras.layers import Dense # 완전 연결 신경망
# minmax scaler도 import해야하지만 위에 있으니까 생략

model = Sequential()
model.add(Dense(40, input_dim=5, activation='relu')) # 첫 은닉층 뉴런 25, 인력 차원 5, 활성화함수 렐루
model.add(Dense(40, activation='relu')) # 첫 은닉층 뉴런 25,  활성화함수 렐루
# 여기선 입력 차원을 입력하지 않음
model.add(Dense(1, activation='linear')) # 출력층
```

```python
model.summary()
# 5 * 25 = 125 + 가중치(25)
# 25 * 25 = 625 + 가중치(25)
# 25 * 1 = 25 + 가중치(1)
```

```python
model.compile(optimizer='adam', loss = 'mean_squared_error') # 경사하강법, 역전파같은 역할
```

```python
epochs_hist = model.fit(X_train,y_train, epochs=20, batch_size=25,verbose = 1,validation_split=0.2)
# 에폭 반복 수, 학습 한번에 들어갈 데이터 수, 훈련중에 정보를 얼마나 보여줄지, 
# 검증분할 => 과대적합을 막기 위해 훈련데이터를 또 검증하기 위한 분할
# batch 증가시 메모리 많이 먹지만 빨라짐
```

```python
epochs_hist = model.fit(X_train,y_train, epochs=100, batch_size=25,verbose = 1,validation_split=0.2)
# 공정하진 않지만, batch 사이즈
```

# 모델검증

```python
epochs_hist.history.keys()
```

```python
plt.plot(epochs_hist.history['loss'])
plt.plot(epochs_hist.history['val_loss'])
plt.title('model loss progress during training')
plt.ylabel('training loss')
plt.xlabel('epoch number')
plt.legend(['training loss','validation loss'])
```
