```python
# np array 기본 
import numpy as np

x = np.array([1., 2., 3.])
y = np.array([2., 4., 6.])

# print(type(x))

# np 연산
print(x + y)
print(x - y)
print(x * y)
print(x / y)
print(type(x))
```

```python
# 원소 접근
import numpy as np
x = np.array([[51,55],[14,19],[0,4]])
print(x)
print(x[0])
print(x[0][0])

# 배열로 접근
for row in x:
    print(x)
    
x = x.flatten() # 1차원 배열로 변환
print(x)
print(x[np.array([0,2,4])]) # 인덱스가 0,2,4인 배열 얻기

print(x>15)
print(x[x>15])
```

```python
# matplotlib

import numpy as np
import matplotlib.pyplot as plt

x= np.arange(0,6,0.1) # 0부터 6까지 0.1 단위로 생성
y = np.sin(x)

#그래프 그리기
plt.plot(x,y)
plt.show()
```

```python
# pyplot 기능 사용하기
import numpy as np
import matplotlib.pyplot as plt

# 데이터 셋
x= np.arange(0,6,0.1)
y1 = np.sin(x)
y2 = np.cos(x)

#그래프 그리기
plt.plot(x,y1,label="sinx")
plt.plot(x,y2,linestyle="--",label="cosx")
plt.xlabel("x") # x축 이름
plt.ylabel("y") # y축 이름
plt.title("sinx and cosx")
plt.legend() # label을 표시
plt.show()
```
